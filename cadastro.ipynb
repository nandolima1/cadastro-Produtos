```python
def adicionar_produto(estoque):
    nome = input("Digite o nome do produto: ")
    quantidade = int(input("Digite a quantidade a ser adicionada: "))
    estoque[nome] = estoque.get(nome, 0) + quantidade

def atualizar_quantidade(estoque):
    nome = input("Digite o nome do produto: ")
    nova_quantidade = int(input("Digite a nova quantidade: "))
    if nome in estoque:
        estoque[nome] = nova_quantidade
    else:
        print(f"Produto {nome} não encontrado no estoque.")

def remover_produto(estoque):
    nome = input("Digite o nome do produto: ")
    if nome in estoque:
        del estoque[nome]
    else:
        print(f"Produto {nome} não encontrado no estoque.")

def exibir_estoque(estoque):
    for nome, quantidade in estoque.items():
        print(f"{nome}: {quantidade}")

def main():
    estoque = {}
    while True:
        print("**Menu**")
        print("1. Adicionar produto")
        print("2. Atualizar quantidade de um produto")
        print("3. Remover produto")
        print("4. Exibir estoque")
        print("5. Sair do programa")

        opcao = int(input("Digite a opção desejada: "))

        if opcao == 1:
            adicionar_produto(estoque)
        elif opcao == 2:
            atualizar_quantidade(estoque)
        elif opcao == 3:
            remover_produto(estoque)
        elif opcao == 4:
            exibir_estoque(estoque)
        elif opcao == 5:
            break
        else:
            print("Opção inválida.")

        continuar = input("Deseja continuar? (S/N) ").upper()
        if continuar != "S":
            break

if __name__ == "__main__":
    main()
```

```python
def adicionar_funcionario(funcionarios):
    nome = input("Digite o nome do funcionário: ")
    if nome in funcionarios:
        print("Funcionário já cadastrado.")
    else:
        funcionarios.add(nome)
        print("Funcionário cadastrado com sucesso.")

def adicionar_visitante(visitantes):
    nome = input("Digite o nome do visitante: ")
    if nome in visitantes:
        print("Visitante já cadastrado.")
    else:
        visitantes.add(nome)
        print("Visitante cadastrado com sucesso.")

def verificar_autorizacao(funcionarios, visitantes):
    nome = input("Digite o nome da pessoa: ")
    if nome in funcionarios:
        print("Acesso autorizado.")
    elif nome in visitantes:
        print("Acesso autorizado (visitante).")
    else:
        print("Acesso negado.")

def remover_autorizacao(funcionarios, visitantes):
    nome = input("Digite o nome da pessoa: ")
    if nome in funcionarios:
        funcionarios.remove(nome)
        print("Autorização de acesso do funcionário removida.")
    elif nome in visitantes:
        visitantes.remove(nome)
        print("Autorização de acesso do visitante removida.")
    else:
        print("Nome não encontrado.")

def main():
    funcionarios = set()
    visitantes = set()
    while True:
        print("**Menu**")
        print("1. Adicionar funcionário autorizado")
        print("2. Adicionar visitante autorizado")
        print("3. Verificar autorização de acesso")
        print("4. Remover autorização de acesso")
        print("5. Sair")

        opcao = int(input("Digite a opção desejada: "))

        if opcao == 1:
            adicionar_funcionario(funcionarios)
        elif opcao == 2:
            adicionar_visitante(visitantes)
        elif opcao == 3:
            verificar_autorizacao(funcionarios, visitantes)
        elif opcao == 4:
            remover_autorizacao(funcionarios, visitantes)
        elif opcao == 5:
            break
        else:
            print("Opção inválida.")

if __name__ == "__main__":
    main()
```
